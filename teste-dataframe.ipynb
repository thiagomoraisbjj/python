```python
df_transformado = df_raw.copy()

df_transformado = df_transformado.drop_duplicates()

df_transformado["valor"] = pd.to_numeric(df_transformado["valor"], errors="coerce")
df_transformado["valor_total"] = pd.to_numeric(df_transformado["valor_total"], errors="coerce")
df_transformado["data_inclusao"] = pd.to_datetime(df_transformado["data_inclusao"], errors="coerce")
df_transformado["data_limite"] = pd.to_datetime(df_transformado["data_limite"], errors="coerce")

df_transformado["dependencia"] = df_transformado["dependencia"].fillna("Titular")

print("\n======= TRANSFORMANDO ========")
print(df_transformado.head())
```
