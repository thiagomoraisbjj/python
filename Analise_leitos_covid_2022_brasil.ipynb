```python
import pandas as pd

url = "https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/LEITOS/2026-04-11/esus-vepi.LeitoOcupacao_2022.csv"

df_sus = pd.read_csv(url)
df_sus.head()
df_sus.info()
```

```python
columns_to_drop = [
    'ocupacaoSuspeitoCli',
    'ocupacaoSuspeitoUti',
    'ocupacaoConfirmadoCli',
    'ocupacaoConfirmadoUti'
]

df_sus = df_sus.drop(columns=columns_to_drop)
print("Colunas foram elimininadas com sucesso:")
df_sus.info()
```

```python
df_analise_sus = df_sus
```

```python
display(df_analise_sus)
```

```python
import pandas as pd

# 1. Converter para datetime (corrigindo a sintaxe)
df_analise_sus['dataNotificacao'] = pd.to_datetime(df_analise_sus['dataNotificacao'])

# 2. Criar uma coluna formatada para exibição (ex: DD/MM/YYYY)
df_analise_sus['data_formatada'] = df_analise_sus['dataNotificacao'].dt.strftime('%d/%m/%Y')

# Exibir as primeiras linhas para conferir
display(df_analise_sus[['dataNotificacao', 'data_formatada']].head())
```

```python
# Usamos errors='ignore' para evitar erro caso a coluna já tenha sido removida anteriormente
df_analise_sus = df_analise_sus.drop(columns=['dataNotificacao'], errors='ignore')

print("Processo de remoção concluído (se a coluna existia, foi removida).")
print("Colunas atuais no DataFrame:", df_analise_sus.columns.tolist())
display(df_analise_sus.head())
```

```python
# Verificar as colunas atuais
print("Colunas no DataFrame df_analise_sus:")
print(df_analise_sus.columns.tolist())

# Exibir uma amostra dos dados
display(df_analise_sus.head())
```

```python
# Converter e formatar _created_at
df_analise_sus['data_criacao'] = pd.to_datetime(df_analise_sus['_created_at']).dt.strftime('%d/%m/%Y')

# Converter e formatar _updated_at
df_analise_sus['data_atualizacao'] = pd.to_datetime(df_analise_sus['_updated_at']).dt.strftime('%d/%m/%Y')

# Remover as colunas originais em inglês
df_analise_sus = df_analise_sus.drop(columns=['_created_at', '_updated_at'], errors='ignore')

print("Colunas traduzidas e formatadas com sucesso!")
display(df_analise_sus[['data_criacao', 'data_atualizacao']].head())
```

```python
display(df_analise_sus.head())
display(df_analise_sus.info())
display(df_analise_sus.describe())
```

```python
# Exibir resumo estatístico das colunas numéricas
print("Resumo Estatístico do DataFrame:")
display(df_analise_sus.describe())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Garantir que a data está em formato datetime para ordenação correta no gráfico
# Criamos uma cópia temporária para não afetar o DF original
df_plot = df_analise_sus.copy()
df_plot['data_dt'] = pd.to_datetime(df_plot['data_formatada'], dayfirst=True)

# Agrupar por data e somar as ocupações
ocupacao_tempo = df_plot.groupby('data_dt')[['ocupacaoCovidUti', 'ocupacaoCovidCli']].sum().sort_index()

# Plotagem
plt.figure(figsize=(12, 6))
plt.plot(ocupacao_tempo.index, ocupacao_tempo['ocupacaoCovidUti'], label='Ocupação UTI COVID', marker='o', markersize=4)
plt.plot(ocupacao_tempo.index, ocupacao_tempo['ocupacaoCovidCli'], label='Ocupação Clínico COVID', marker='s', markersize=4)

plt.title('Ocupação de Leitos COVID ao Longo do Tempo (Brasil - 2022)')
plt.xlabel('Data de Notificação')
plt.ylabel('Total de Leitos Ocupados')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Selecionar colunas de ocupação e verificar valores nulos
ocupacao_cols = ['ocupacaoCovidUti', 'ocupacaoCovidCli', 'ocupacaoHospitalarUti', 'ocupacaoHospitalarCli']

print("Valores ausentes por coluna de ocupação:")
display(df_analise_sus[ocupacao_cols].isna().sum())

# Verificar também se existem valores zero (que em alguns contextos podem indicar falta de preenchimento)
print("\nQuantidade de registros com valor zero:")
display((df_analise_sus[ocupacao_cols] == 0).sum())
```

```python
# Calcular a correlação entre ocupação de UTI e leitos clínicos
correlation = df_analise_sus['ocupacaoCovidUti'].corr(df_analise_sus['ocupacaoCovidCli'])

print(f"A correlação de Pearson entre ocupação de UTI e Clínico COVID é: {correlation:.4f}")

# Exibir a matriz de correlação completa para as colunas de ocupação
print("\nMatriz de Correlação Completa:")
display(df_analise_sus[['ocupacaoCovidUti', 'ocupacaoCovidCli', 'ocupacaoHospitalarUti', 'ocupacaoHospitalarCli']].corr())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_analise_sus, x='ocupacaoCovidCli', y='ocupacaoCovidUti', alpha=0.5)

plt.title('Dispersão: Ocupação de Leitos Clínicos vs. UTI (COVID)')
plt.xlabel('Ocupação de Leitos Clínicos')
plt.ylabel('Ocupação de Leitos UTI')
plt.grid(True, linestyle='--', alpha=0.6)

# Adicionar uma linha de tendência para melhor visualização
sns.regplot(data=df_analise_sus, x='ocupacaoCovidCli', y='ocupacaoCovidUti', scatter=False, color='red')

plt.show()
```

```python
# Identificar os 10 hospitais com maior média de ocupação de UTI COVID
hospitais_pressao = df_analise_sus.groupby(['cnes', 'municipio', 'estado'])['ocupacaoCovidUti'].mean().sort_values(ascending=False).reset_index()

print("Top 10 Hospitais com maior média de ocupação de UTI COVID:")
display(hospitais_pressao.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Preparar os dados para o gráfico (Top 10)
top_10_hospitais = hospitais_pressao.head(10).copy()
top_10_hospitais['hospital_info'] = top_10_hospitais['municipio'] + " (" + top_10_hospitais['estado'] + ") - CNES: " + top_10_hospitais['cnes']

plt.figure(figsize=(12, 7))
sns.barplot(data=top_10_hospitais, x='ocupacaoCovidUti', y='hospital_info', palette='Reds_r')

plt.title('Top 10 Hospitais com Maior Média de Ocupação de UTI COVID (2022)')
plt.xlabel('Média de Leitos de UTI Ocupados')
plt.ylabel('Hospital / Localização')
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
